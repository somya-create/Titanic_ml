# Linear Model 

```python
ld_train, ld_test = train_test_split(titanic_simple, test_size = 0.2,random_state=2)
```

```python
lm=LinearRegression()
```

```python
x_train=ld_train.drop(["Survived"],1)
y_train=ld_train["Survived"]
x_test=ld_test.drop(["Survived"],1)
y_test=ld_test["Survived"]
```

```python
lm.fit(x_train,y_train)
```

```python
p_test=lm.predict(x_test)

residual=p_test-y_test

rmse_lm=np.sqrt(np.dot(residual,residual)/len(p_test))

rmse_lm
```

```python
coefs=lm.coef_

features=x_train.columns

list(zip(features,coefs))
```
